```python
import tensorflow as tf 
import pandas as pd 
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras import layers,models
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
# normalize to [0,1]
X_train = X_train.astype("float32") / 255.0   
X_test  = X_test.astype("float32")  / 255.0
```

```python
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test  = tf.keras.utils.to_categorical(y_test,  10)
```

```python
inputs = tf.keras.Input(shape=(32, 32, 3))
x = inputs
```

```python
# Conv Block 1 — detect basic features
x = layers.Conv2D(32, 3, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.Conv2D(32, 3, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D(2)(x)        # 32x32 → 16x16
x = layers.Dropout(0.25)(x)
```

```python
# Conv Block 2 — detect shapes
x = layers.Conv2D(64, 3, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.Conv2D(64, 3, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D(2)(x)        # 16x16 → 8x8
x = layers.Dropout(0.25)(x)
```

```python
# Conv Block 3 — detect complex patterns
x = layers.Conv2D(128, 3, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.Conv2D(128, 3, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D(2)(x)        # 8x8 → 4x4
x = layers.Dropout(0.25)(x)
```

```python
x = layers.GlobalAveragePooling2D()(x)   # 4x4x128 → 128
x = layers.Dense(256)(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(10, activation='softmax')(x)
```

```python
model = tf.keras.Model(inputs, outputs)
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50,
    batch_size=64,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_accuracy', patience=8, restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss', factor=0.5, patience=4
        )
    ]
)
```

```python
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {acc:.4f}")
```
